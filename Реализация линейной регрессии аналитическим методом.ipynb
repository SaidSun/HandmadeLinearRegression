```python
import math
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
import pandas as pd
```

```python
def func(a, b, t, e):
    x = []
    for i in range(len(t)):
        x.append(a*t[i]**2 + b*t[i] + e[i])
    return x
```

```python
a = 0.8
b = 20
t = np.arange(0, 250, 0.5)
t_2 = [i**2 for i in t]
ones = [1 for i in t]
l = 10000
er = np.random.exponential(l, 500)
x = np.array(func(a, b, t, er))
C_T = np.array([t_2, t, ones])
C = C_T.T
print(er, x)
plt.title('График исходной выборки')
plt.plot(t, x)
```

```python
theta = np.dot(np.dot(np.linalg.inv(np.dot(C_T, C)), C_T), x)
```
